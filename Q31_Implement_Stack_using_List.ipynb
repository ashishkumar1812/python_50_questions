```python
class Stack:
    def __init__(self):
        """
        Initialize an empty stack.
        """
        print("This class implements a stack using a list (LIFO).")
        self.items = []

    def push(self, item):
        """
        Add an element to the top of the stack.

        Args:
            item: Any value to be added to the stack.
        """
        self.items.append(item)
        print(f"Pushed: {item}")

    def pop(self):
        """
        Remove and return the top element of the stack.

        Returns:
            The top element if stack is not empty, otherwise None.
        """
        if self.items:
            popped_item = self.items.pop()
            print(f"Popped: {popped_item}")
            return popped_item
        else:
            print("Stack is empty. Cannot pop.")
            return None

    def peek(self):
        """
        Return the top element without removing it.

        Returns:
            The top element if stack is not empty, otherwise None.
        """
        if self.items:
            top_item = self.items[-1]
            print(f"Top element: {top_item}")
            return top_item
        else:
            print("Stack is empty. Nothing to peek.")
            return None

    def is_empty(self):
        """
        Check if the stack is empty.

        Returns:
            bool: True if stack is empty, False otherwise.
        """
        return len(self.items) == 0


# main execution
if __name__ == "__main__":
    stack = Stack()

    # push elements
    stack.push(10)
    stack.push(20)
    stack.push(30)

    # peek at top
    stack.peek()  # Expected Output: 30

    # pop elements
    stack.pop()   # Expected Output: 30
    stack.peek()  # Expected Output: 20
    stack.pop()   # Expected Output: 20
    stack.pop()   # Expected Output: 10
    stack.pop()   # Expected Output: None (stack empty)
```
