```python
class ArmstrongNumberChecker:
    def __init__(self):
        print("This class checks if a number is an Armstrong number.")

    def is_armstrong(self, number: int) -> bool:
        """
        Check if a number is an Armstrong number.

        Args:
            number (int): The number to check

        Returns:
            bool: True if the number is an Armstrong number, False otherwise

        Raises:
            TypeError: If input is not an integer
            ValueError: If number is negative
        """
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")
        if number < 0:
            raise ValueError("Number must be non-negative.")

        num_str = str(number)
        n = len(num_str)
        sum_of_powers = sum(int(digit) ** n for digit in num_str)

        return sum_of_powers == number


# main execution
if __name__ == "__main__":
    try:
        number_input = input("Enter a non-negative integer: ").strip()

        if not number_input.isdigit():
            raise TypeError("Input must be a non-negative integer.")
        number = int(number_input)

        obj = ArmstrongNumberChecker()
        result = obj.is_armstrong(number)

        if result:
            print(f"{number} is an Armstrong number.")
        else:
            print(f"{number} is NOT an Armstrong number.")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
