```python
# Convert Decimal to Binary
num = 10
if num == 0:
    print("0")

binary = ""
while num > 0:
    binary = str(num % 2) + binary
    num //= 2
print(binary)
```

```python
class DecimalToBinary:
    def __init__(self):
        print("This class converts a decimal number to binary.")

    def decimal_to_binary(self, number: int) -> str:
        """
        This method converts a decimal number to binary.
        
        Args:
            number (int): A non-negative integer.
        
        Returns:
            str: Binary representation of the number.
        
        Raises:
            TypeError: If input is not an integer.
            ValueError: If input is negative.
        """
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")
        if number < 0:
            raise ValueError("Input must be a non-negative integer.")

        if number == 0:
            return "0"

        binary = ""
        while number > 0:
            binary = str(number % 2) + binary
            number //= 2

        return binary


# main execution
if __name__ == "__main__":
    try:
        user_input = int(input("Enter a decimal number: "))
        obj = DecimalToBinary()
        result = obj.decimal_to_binary(user_input)
        print(f"Binary representation of {user_input} is: {result}")
    except TypeError as e:
        print("TypeError:", e)
    except ValueError as e:
        print("ValueError:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
