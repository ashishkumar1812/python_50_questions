```python
class CommonElementsFinder:
    def __init__(self):
        print("This class finds elements common in three lists.")

    def find_common(self, list1: list, list2: list, list3: list) -> list:
        """
        Find elements that are common in three lists.

        Args:
            list1 (list): First list
            list2 (list): Second list
            list3 (list): Third list

        Returns:
            list: List of elements common to all three lists

        Raises:
            TypeError: If any of the inputs are not lists
        """
        if not isinstance(list1, list) or not isinstance(list2, list) or not isinstance(list3, list):
            raise TypeError("All inputs must be lists.")

        common_elements = []
        for item in list1:
            if item in list2 and item in list3 and item not in common_elements:
                common_elements.append(item)

        return common_elements


# main execution
if __name__ == "__main__":
    try:
        raw_list1 = input("Enter the first list (e.g., [1,2,3,4,5]): ").strip()
        list1 = eval(raw_list1)

        raw_list2 = input("Enter the second list (e.g., [3,4,5,6,7]): ").strip()
        list2 = eval(raw_list2)

        raw_list3 = input("Enter the third list (e.g., [5,6,7,8,9]): ").strip()
        list3 = eval(raw_list3)

        obj = CommonElementsFinder()
        result = obj.find_common(list1, list2, list3)

        print("Common elements in all three lists:")
        print(result)

    except (TypeError, SyntaxError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
