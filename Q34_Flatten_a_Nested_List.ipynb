```python
class NestedListFlattener:
    def __init__(self):
        print("This class flattens a nested list into a single list of integers.")

    def flatten(self, nested_list: list) -> list:
        """
        Flatten a nested list into a single-level list.

        Args:
            nested_list (list): List containing integers or nested lists

        Returns:
            list: Flattened list with all integers

        Raises:
            TypeError: If input is not a list or contains non-integer/non-list elements
        """
        if not isinstance(nested_list, list):
            raise TypeError("Input must be a list.")

        flat_list = []

        for item in nested_list:
            if isinstance(item, list):
                flat_list.extend(self.flatten(item))  # recursive call
            elif isinstance(item, int):
                flat_list.append(item)
            else:
                raise TypeError("List can only contain integers or nested lists.")

        return flat_list


# main execution
if __name__ == "__main__":
    try:
        raw_input_list = input("Enter a nested list using Python list syntax (e.g., [1, [2, [3, 4]], 5]): ").strip()
        nested_list = eval(raw_input_list)  # careful: assumes trusted input

        obj = NestedListFlattener()
        flattened = obj.flatten(nested_list)

        print("Flattened list:")
        print(flattened)

    except (TypeError, SyntaxError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
