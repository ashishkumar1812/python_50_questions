```python
class SumOfList:
    def __init__(self):
        print("This class calculates the sum of a list using recursion.")

    def recursive_sum(self, numbers: list) -> int | float:
        """
        Calculate the sum of a list of numbers using recursion.

        Args:
            numbers (list): List containing numeric values

        Returns:
            int | float: Sum of the list elements

        Raises:
            TypeError: If input is not a list or contains non-numeric values
            ValueError: If the list is empty
        """
        # validate input type
        if not isinstance(numbers, list):
            raise TypeError("Input must be a list.")

        if not numbers:
            raise ValueError("List must not be empty.")

        # validate elements
        for item in numbers:
            if not isinstance(item, (int, float)):
                raise TypeError("List elements must be numbers.")

        # base case
        if len(numbers) == 1:
            return numbers[0]

        # recursive case
        return numbers[0] + self.recursive_sum(numbers[1:])


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter numbers separated by spaces: ").strip()

        if not user_input:
            raise ValueError("Input must not be empty.")

        numbers = []
        for value in user_input.split():
            numbers.append(float(value))

        obj = SumOfList()
        result = obj.recursive_sum(numbers)

        print("Sum of the list:", result)

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
