```python
class VariableSwapper:
    def __init__(self):
        print("This class swaps two variables without using a temporary variable.")

    def swap(self, a: int, b: int) -> tuple:
        """
        Swap two numbers without using a temporary variable.

        Args:
            a (int): First number
            b (int): Second number

        Returns:
            tuple: Swapped values (a, b)

        Raises:
            TypeError: If inputs are not integers
        """
        if not isinstance(a, int) or not isinstance(b, int):
            raise TypeError("Both inputs must be integers.")

        # Swap using arithmetic operations
        a = a + b
        b = a - b
        a = a - b

        return a, b


# main execution
if __name__ == "__main__":
    try:
        a_input = input("Enter the value of a: ").strip()
        b_input = input("Enter the value of b: ").strip()

        if not a_input.isdigit() or not b_input.isdigit():
            raise TypeError("Both inputs must be integers.")

        a = int(a_input)
        b = int(b_input)

        obj = VariableSwapper()
        a, b = obj.swap(a, b)

        print(f"After swapping:\na = {a}\nb = {b}")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
