```python
def largest_element_in_list(list1: list[float | int]) -> float:
    """
    Returns the largest element in a numeric list.

    Args:
        list1 (list[float | int]): A list containing numeric values (int or float)

    Returns:
        float: The largest element in the list

    Raises:
        TypeError: If the input is not a list or contains non-numeric items
        ValueError: If the list is empty
    """
    if not isinstance(list1, list):
        raise TypeError("Input must be a list")
    
    if not list1:
        raise ValueError("List must not be empty")
    
    for item in list1:
        if not isinstance(item, (int, float)):
            raise TypeError("List elements must be numbers")
        
    largest_element = list1[0]
    for i in list1[1:]:
        if i > largest_element:
            largest_element = i

    return largest_element


if __name__ == "__main__":
    try:
        raw_input = input("Enter numbers separated by commas: ")
        list1 = list(map(float, raw_input.split(",")))  # float handles both int & float
        result = largest_element_in_list(list1)
        print("Largest element:", result)

    except (TypeError, ValueError) as error:
        print("Error:", error)

    except Exception as e:
        print("Unexpected error:", e)
```
