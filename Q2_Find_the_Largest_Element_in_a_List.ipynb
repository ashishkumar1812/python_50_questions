```python
class LargestElement:
    def __init__(self):
        print("This class finds the largest element in a numeric list.")

    def largest_element_in_list(self, list1: list[float | int]) -> float:
        """
        This method returns the largest element in a numeric list.

        Args:
            list1 (list[float | int]): List containing int or float values

        Returns:
            float: Largest element in the list

        Raises:
            TypeError: If input is not a list or contains non-numeric values
            ValueError: If the list is empty
        """
        if not isinstance(list1, list):
            raise TypeError("Input must be a list.")

        if not list1:
            raise ValueError("List must not be empty.")

        for item in list1:
            if not isinstance(item, (int, float)):
                raise TypeError("List elements must be numbers.")

        largest_element = list1[0]
        for value in list1[1:]:
            if value > largest_element:
                largest_element = value

        return largest_element


# main execution
if __name__ == "__main__":
    try:
        raw_input = input("Enter numbers separated by spaces: ")
        numbers = list(map(float, raw_input.split(" ")))  # handles int & float

        obj = LargestElement()
        result = obj.largest_element_in_list(numbers)

        print("Largest element:", result)

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected error:", e)
```
