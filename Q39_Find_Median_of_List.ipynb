```python
class MedianFinder:
    def __init__(self):
        print("This class calculates the median of a list of numbers.")

    def find_median(self, numbers: list) -> float:
        """
        Calculate the median of a list of numbers.

        Args:
            numbers (list): List of integers or floats

        Returns:
            float: Median value

        Raises:
            TypeError: If input is not a list or contains non-numeric elements
            ValueError: If list is empty
        """
        if not isinstance(numbers, list):
            raise TypeError("Input must be a list.")
        if not numbers:
            raise ValueError("List cannot be empty.")
        for item in numbers:
            if not isinstance(item, (int, float)):
                raise TypeError("All elements must be numbers.")

        sorted_numbers = sorted(numbers)
        n = len(sorted_numbers)
        mid = n // 2

        if n % 2 == 1:  # odd number of elements
            return float(sorted_numbers[mid])
        else:  # even number of elements
            return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2


# main execution
if __name__ == "__main__":
    try:
        raw_input_list = input("Enter numbers separated by spaces: ").strip()
        numbers = [float(x) for x in raw_input_list.split()]

        obj = MedianFinder()
        median = obj.find_median(numbers)

        print(f"The median of the list is: {median}")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
