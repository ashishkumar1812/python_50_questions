```python
class FirstNonRepeatedCharFinder:
    def __init__(self):
        print("This class finds the first non-repeated character in a string.")

    def find_first_non_repeated(self, string: str) -> str | None:
        """
        Return the first non-repeated character in a string.

        Args:
            string (str): The string to search

        Returns:
            str | None: The first non-repeated character, or None if all characters repeat

        Raises:
            TypeError: If input is not a string
        """
        if not isinstance(string, str):
            raise TypeError("Input must be a string.")

        char_count = {}
        for char in string:
            char_count[char] = char_count.get(char, 0) + 1

        for char in string:
            if char_count[char] == 1:
                return char

        return None  # all characters repeated


# main execution
if __name__ == "__main__":
    try:
        string_input = input("Enter a string: ").strip()

        obj = FirstNonRepeatedCharFinder()
        result = obj.find_first_non_repeated(string_input)

        if result:
            print(f"The first non-repeated character is: '{result}'")
        else:
            print("All characters are repeated; no non-repeated character found.")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
