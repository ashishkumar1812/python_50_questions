```python
def count_vowels(input_string: str) -> dict:
    """
    This function counts vowels in a string.
    Args:
        Takes string as an input.
    Returns:
        returns a dictionary
    Raises:
        ValueError: checks if string is empty
        TypeError: input must be a string
    """
    
    if not isinstance(input_string,str):
        raise TypeError("Input must be a string")
    input_string = input_string.strip().lower()
    if not input_string:
        raise ValueError("String must not be empty")
    
    vowels = ["a", "e", "i", "o", "u"]
    result = {}
    for ch in input_string:
        if ch in vowels:
            result[ch] = result.get(ch, 0) + 1
    
    for k, v in result.items():
        print(f"{k} = {v}")
    return result

if __name__ == "__main__":
    try:
        input_string = input("Enter a string: ")
        input_string = input_string.strip().lower()
        output = count_vowels(input_string)
        print(output)
    except (TypeError,ValueError) as error:
        print("string should not empty!",error)
    except Exception as error:
        print("something went wrong",error)
```
