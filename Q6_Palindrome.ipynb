```python
name = "nitin"
palin = name[::-1]
if palin == name:
    print("yes it is palindrome")
else:
    print("its not")
```

```python
def palindrome(input_string: str) -> bool:
    """
    Function checks if a string is palindrome.
    Args:
        Takes string as an input.
    Return:
        Boolean: True or False
    Raises:
        ValueError: string must not empty.
        TypeError: input must be a string.
    """
    # input check
    if not isinstance(input_string,str):
        raise TypeError("Input must be a string")
    # removes spaces
    input_string = input_string.strip().lower()
    # empty string check
    if not input_string:
        raise ValueError("Input must not empty")
        
    text = input_string[::-1]
    if text == input_string:
        return True
    else:
        return False

if __name__ == "__main__":
    try:
        input_string = input("Enter a string")
        input_string = input_string.strip().lower()
        if palindrome(input_string):
            print(f"{input_string} is a palindrome")
        else:
            print(f"{input_string} is not a palindrome")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print("something went wrong",e)
        
```
