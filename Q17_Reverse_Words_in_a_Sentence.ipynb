```python
class ReverseSentence:
    def __init__(self):
        print("This is ReverseSentence class")

    def reverse_words(self, user_input: str) -> str:
        """
        Reverses the order of words in a given sentence.

        Args:
            user_input (str): Input sentence

        Returns:
            str: Sentence with words in reverse order

        Raises:
            TypeError: If input is not a string
            ValueError: If input is empty
        """
        if not isinstance(user_input, str):
            raise TypeError("Input must be a string.")
        user_input = user_input.strip()
        if not user_input:
            raise ValueError("Input must not be empty.")
        list1 = user_input.split()
        result = " ".join(list1[::-1])
        return result

if __name__ == "__main__":
    try:
        user_input = input("Enter a sentence: ")
        obj = ReverseSentence()
        result = obj.reverse_words(user_input)
        print(f"Reverse of given sentence is : {result} ")
    except TypeError as e:
        print(e)
    except ValueError as e:
        print("This is Value Error = ",e)
    except Exception as e:
        print("something went wrong.",e)
```
