```python
def second_largest_element(user_input:list) -> float:
    """
    This function finds second largest element in a given list.
    Args:
        Takes list as a input.
    Returns:
        Integer
    Raises:
        TypeError: input must be a list.
        ValueError: list must not be empty
    """
    # list check (if it is a list or not)
    if not isinstance(user_input,list):
        raise TypeError("input must be a list")
    # empty list check
    if not user_input:
        raise ValueError("input must not be empty")
    # all elements of list should be an integer
    for i in user_input:
        if not isinstance(i, (int,float)):
            raise TypeError("All element must be a number")
    #list length should be at least 2
    if len(user_input) < 2:
        raise ValueError("List must contain at least two elements")
    # logic
    largest = user_input[0]
    second_largest = None
    for item in user_input:
        if item > largest:
            second_largest = largest
            largest = item
        elif largest > item > second_largest:
            second_largest = item
    return second_largest

if __name__ == "__main__":
    try:
        user_input = list(map(float, input("Enter elements separated by spaces: ").split()))
        if not user_input:
            raise ValueError("list can not be empty")
        result = second_largest_element(user_input)
        print(f"'{result}' is the second largest element in the given list.")
    except TypeError as e:
        print("list must be an int or float",e)

    except Exception as e:
        print(e)

```
