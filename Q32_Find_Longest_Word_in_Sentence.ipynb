```python
class LongestWordFinder:
    def __init__(self):
        print("This class finds the longest word in a given sentence.")

    def find_longest_word(self, sentence: str) -> str:
        """
        Find and return the longest word in the sentence.

        Args:
            sentence (str): Input sentence

        Returns:
            str: Longest word in the sentence

        Raises:
            TypeError: If input is not a string
            ValueError: If sentence is empty or has no words
        """
        if not isinstance(sentence, str):
            raise TypeError("Input must be a string.")

        words = sentence.strip().split()
        if not words:
            raise ValueError("The sentence must contain at least one word.")

        longest_word = words[0]
        for word in words[1:]:
            if len(word) > len(longest_word):
                longest_word = word

        return longest_word


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a sentence: ").strip()

        obj = LongestWordFinder()
        longest = obj.find_longest_word(user_input)

        print(f"The longest word in the sentence is: '{longest}'")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
