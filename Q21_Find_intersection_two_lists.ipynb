```python
class ListIntersection:
    def __init__(self):
        print("This class finds the intersection of two lists without using set operations.")

    def intersection_of_lists(self, list1: list, list2: list) -> list:
        """
        This method returns the intersection of two lists.

        Args:
            list1 (list): First input list
            list2 (list): Second input list

        Returns:
            list: List containing common elements from both lists

        Raises:
            TypeError: If inputs are not lists
        """
        # validate input types
        if not isinstance(list1, list) or not isinstance(list2, list):
            raise TypeError("Both inputs must be lists.")

        intersection_list = []

        for item in list1:
            if item in list2 and item not in intersection_list:
                intersection_list.append(item)

        return intersection_list


# main execution
if __name__ == "__main__":
    try:
        # taking input from user
        input1 = input("Enter first list elements separated by spaces: ")
        input2 = input("Enter second list elements separated by spaces: ")

        list1 = input1.split()
        list2 = input2.split()

        obj = ListIntersection()
        result = obj.intersection_of_lists(list1, list2)

        print("Intersection of the two lists:")
        print(result)

    except TypeError as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
