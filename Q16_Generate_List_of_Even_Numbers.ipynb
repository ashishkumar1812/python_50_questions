```python
class EvenNumbers:
    def __init__(self):
        print("This is EvenNumber class.")
    def list_of_even_numbers(self, number : int) -> list:
        """
        This method finds even numbers in a range from 2 to given number.
        Args:
            int: Number
        Returns:
            list: list of Even numbers.
        Raises:
            TypeError and ValueError
        """
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")
        if number < 2:
            raise ValueError("Input must be greater than or equal to 2")
        list1 = []
        for num in range(2, number+1):
            if num % 2 == 0:
                list1.append(num)
        
        return list1

if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number greater than 2: "))
        obj = EvenNumbers()
        result = obj.list_of_even_numbers(user_input)
        print(f"Even number list: {result}")
    except TypeError as e:
        print("Input must be a integer.")
    except ValueError as e:
        print("This is Value Error = ",e)
    except Exception as e:
        print("something went wrong.",e)
```
