```python
# Calculate Power without Using ** Operator
number = 2
power = 3
result = 1
for _ in range(power):
    result = result * number
print(result)
```

```python
class Power:
    def __init__(self):
        print("This is Power class")
        print("Calculate Power without Using ** Operator")

    def calculate_power(self, number, power : int) -> int:
        """
        This method calculates power of a number.
        Args:
            It takes two integers: Number and power.
        Return:
            int: Power of an interger.
        raises:
            ValueError and TypeError
        """
        if not isinstance(number, int) or isinstance(power, int):
            raise TypeError("Enter integer only for number and power.")
        if power < 0:
            raise ValueError("number or power should not be less than 0")
        result = 1
        for _ in range(power):
            result = result * number
        return result

if __name__ == "__main__":
    try:
        number = int(input("Enter a number: "))
        power = int(input("Enter power for the number: "))
        obj = Power()
        result = obj.calculate_power(number,power)
        print(f"Power of {number} is {result}")
    except TypeError as e:
        print("Input must be a integer.")
    except ValueError as e:
        print("Value Error",e)
    except Exception as e:
        print("something went wrong.",e)
```
