```python
number = 10
if number < 2:
    print("not prime")
else:
    for i in range(2,number):
        if number % i == 0:
            print("not prime")
            break
    else:
        print("prime")
```

```python
def prime_check(number: int) -> bool:
    """
    Check whether a given number is prime.

    Args:
        number (int): Number to be checked.

    Returns:
        bool: True if the number is prime, False otherwise.

    Raises:
        TypeError: If input is not an integer.
        ValueError: if number is less than 2.
    """
    # validate input an integer
    if not isinstance(number,int):
        raise TypeError("Input is not Int")

    # number should be greater than 1
    if number < 2:
        raise ValueError("Number is less than 2")
    # check divisibility upto square root of the number.
    for i in range(2,int(number ** .5) + 1):
        if number % i == 0:
            return False
    else:
        return True

if __name__ == "__main__":
    try:
        # taking input from user
        user_input = input("Enter a number: ")
        #validate and convert the number into integer
        if not user_input.strip().isdigit():
            raise TypeError("Input must be an integer")
        number = int(user_input)
        # Calling the function
        if prime_check(number):
            print(f"{number} is a prime number")
        else:
            print(f"{number} is a not prime number")

    except (TypeError,ValueError) as Error:
        print("number should be int",Error)
    except Exception as Error:
        print("something went wrong",Error)
```
