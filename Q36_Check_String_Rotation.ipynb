```python
class StringRotationChecker:
    def __init__(self):
        print("This class checks if two strings are rotations of each other.")

    def are_rotations(self, str1: str, str2: str) -> bool:
        """
        Check if str2 is a rotation of str1.

        Args:
            str1 (str): First string
            str2 (str): Second string

        Returns:
            bool: True if str2 is a rotation of str1, False otherwise

        Raises:
            TypeError: If inputs are not strings
        """
        if not isinstance(str1, str) or not isinstance(str2, str):
            raise TypeError("Both inputs must be strings.")

        if len(str1) != len(str2):
            return False

        # If str2 is a substring of str1+str1, it is a rotation
        return str2 in (str1 + str1)


# main execution
if __name__ == "__main__":
    try:
        str1 = input("Enter the first string: ").strip()
        str2 = input("Enter the second string: ").strip()

        obj = StringRotationChecker()
        result = obj.are_rotations(str1, str2)

        if result:
            print(f"'{str2}' is a rotation of '{str1}'.")
        else:
            print(f"'{str2}' is NOT a rotation of '{str1}'.")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
