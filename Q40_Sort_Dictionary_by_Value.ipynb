```python
class DictionarySorter:
    def __init__(self):
        print("This class sorts a dictionary by its values in ascending order.")

    def sort_by_value(self, input_dict: dict) -> dict:
        """
        Sort a dictionary by its values.

        Args:
            input_dict (dict): Dictionary to sort

        Returns:
            dict: Dictionary sorted by values in ascending order

        Raises:
            TypeError: If input is not a dictionary or contains non-comparable values
        """
        if not isinstance(input_dict, dict):
            raise TypeError("Input must be a dictionary.")

        try:
            sorted_items = sorted(input_dict.items(), key=lambda item: item[1])
        except TypeError:
            raise TypeError("All values must be comparable (int or float).")

        return dict(sorted_items)


# main execution
if __name__ == "__main__":
    try:
        raw_input_dict = input(
            "Enter dictionary using Python syntax (e.g., {'apple': 5, 'banana': 2}): "
        ).strip()
        input_dict = eval(raw_input_dict)  # assumes trusted input

        obj = DictionarySorter()
        sorted_dict = obj.sort_by_value(input_dict)

        print("Dictionary sorted by values:")
        print(sorted_dict)

    except (TypeError, SyntaxError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
