```python
class ListToDictConverter:
    def __init__(self):
        print("This class converts a list of tuples into a dictionary.")

    def convert(self, tuples_list: list) -> dict:
        """
        Convert a list of tuples into a dictionary.

        Args:
            tuples_list (list): List of tuples, each containing (key, value)

        Returns:
            dict: Dictionary with keys and values from the list of tuples

        Raises:
            TypeError: If input is not a list or tuples are not of length 2
        """
        if not isinstance(tuples_list, list):
            raise TypeError("Input must be a list.")

        result_dict = {}
        for item in tuples_list:
            if not isinstance(item, tuple) or len(item) != 2:
                raise TypeError("Each item must be a tuple of length 2 (key, value).")
            key, value = item
            result_dict[key] = value

        return result_dict


# main execution
if __name__ == "__main__":
    try:
        raw_input_list = input(
            "Enter list of tuples using Python syntax (e.g., [('a', 1), ('b', 2)]): "
        ).strip()
        tuples_list = eval(raw_input_list)  # assumes trusted input

        obj = ListToDictConverter()
        result = obj.convert(tuples_list)

        print("Converted dictionary:")
        print(result)

    except (TypeError, SyntaxError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
