```python
def factorial(number):
    if number == 0 or number == 1:
        return 1
    return number * factorial(number-1)

factorial(7)
```

```python
class Factorial:
    def __init__(self):
        print("This class gives factorial of a number using recursion.")

    def factorial(self,number):
        """
        This method takes an integer as an input and returns its factorial.
        Args:
            number = A positive integer.
        Returns:
            Factorial of a number.
        Raises:
            TypeError and ValueError
        """
        if not isinstance(number,int):
            raise TypeError("TypeError")
        if number < 0:
            raise ValueError("Factorial is not defined for negative numbers")
        if number == 0 or number == 1:
            return 1
        return number * self.factorial(number-1)

if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number: "))
        obj = Factorial()
        result = obj.factorial(user_input)
        print(f"factorial of {user_input} is {result}")
        
    except TypeError as e:
        print("TypeError - Input must be a integer.")
    except ValueError as e:
        print("ValueError - Input must be a positive integer.")
    except Exception as e:
        print(e)
```
