```python
def prime_check(number: int) -> bool:
    """
    Check whether a given number is prime.

    Args:
        number (int): Number to be checked.

    Returns:
        bool: True if the number is prime, False otherwise.

    Raises:
        TypeError: If input is not an integer.
        ValueError: if number is less than 2.
    """
    # validate input an integer
    if not isinstance(number,int):
        raise TypeError("Input is not Int")

    # number should be greater than 1
    if number < 2:
        raise ValueError("Number is less than 2")
    # check divisibility upto square root of the number.
    for i in range(2,int(number ** .5) + 1):
        if number % i == 0:
            return False
    else:
        return True

if __name__ == "__main__":
    try:
        prime_number = []
        lower_limit = input("Enter the lower limit: ")
        upper_limit = input("Enter the upper limit: ")
        if not lower_limit.strip().isdigit():
            raise TypeError("Input must be an integer (no decimals or letters).")
        if not upper_limit.strip().isdigit():
            raise TypeError("Input must be an integer (no decimals or letters).")
        lower_limit = int(lower_limit)
        upper_limit = int(upper_limit)            
        for i in range(lower_limit,upper_limit + 1):
            if prime_check(i):
                print(f"{i} is a prime number.")
                prime_number.append(i)
        print(prime_number)
        
    except (TypeError, ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
    
```
