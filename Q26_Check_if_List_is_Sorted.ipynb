```python
class CheckListSorted:
    def __init__(self):
        print("This class checks if a list is sorted in ascending order.")

    def is_sorted(self, lst: list) -> bool:
        """
        Check whether the given list is sorted in ascending order.

        Args:
            lst (list): List of numbers

        Returns:
            bool: True if list is sorted ascending, False otherwise

        Raises:
            TypeError: If input is not a list or contains non-numeric values
        """
        if not isinstance(lst, list):
            raise TypeError("Input must be a list.")

        for item in lst:
            if not isinstance(item, (int, float)):
                raise TypeError("All elements in the list must be numbers.")

        for i in range(len(lst) - 1):
            if lst[i] > lst[i + 1]:
                return False
        return True


# main execution
if __name__ == "__main__":
    try:
        raw_input_list = input("Enter list of numbers separated by space: ").strip()
        numbers = [float(x) for x in raw_input_list.split()]

        obj = CheckListSorted()
        result = obj.is_sorted(numbers)

        if result:
            print("The list is sorted in ascending order.")
        else:
            print("The list is NOT sorted in ascending order.")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
