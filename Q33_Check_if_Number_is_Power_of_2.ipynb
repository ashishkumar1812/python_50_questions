```python
class PowerOfTwoChecker:
    def __init__(self):
        print("This class checks whether a number is a power of 2.")

    def is_power_of_two(self, number: int) -> bool:
        """
        Check if the given number is a power of 2.

        Args:
            number (int): Number to check

        Returns:
            bool: True if number is power of 2, False otherwise

        Raises:
            TypeError: If input is not an integer
            ValueError: If number is less than 1
        """
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")
        if number < 1:
            raise ValueError("Number must be greater than 0.")

        # A number is power of 2 if it has only one bit set
        return (number & (number - 1)) == 0


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a positive integer: ").strip()

        if not user_input.isdigit():
            raise TypeError("Input must be a positive integer.")

        number = int(user_input)

        obj = PowerOfTwoChecker()
        result = obj.is_power_of_two(number)

        if result:
            print(f"{number} is a power of 2.")
        else:
            print(f"{number} is NOT a power of 2.")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
