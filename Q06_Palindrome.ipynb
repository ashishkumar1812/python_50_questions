```python
name = "nitin"
palin = name[::-1]
if palin == name:
    print("yes it is palindrome")
else:
    print("its not")
```

```python
class Palindrome:
    def __init__(self):
        print("This class checks whether a string is a palindrome.")

    def palindrome_check(self, input_string: str) -> bool:
        """
        This method checks if a string is a palindrome.

        Args:
            input_string (str): String to be checked.

        Returns:
            bool: True if palindrome, False otherwise.

        Raises:
            TypeError: If input is not a string.
            ValueError: If input string is empty.
        """
        # type check
        if not isinstance(input_string, str):
            raise TypeError("Input must be a string.")

        # remove spaces and normalize case
        input_string = input_string.strip().lower()

        # empty check
        if not input_string:
            raise ValueError("Input must not be empty.")

        # palindrome logic
        reversed_string = input_string[::-1]
        return input_string == reversed_string


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a string: ")

        obj = Palindrome()
        result = obj.palindrome_check(user_input)

        if result:
            print(f"'{user_input}' is a palindrome.")
        else:
            print(f"'{user_input}' is not a palindrome.")

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
