```python
class Anagram:
    def __init__(self):
        print("This class checks if two strings are anagram or not")

    def Anagram_check(self,input1,input2: str) -> str:
        
        """
        Checks whether two strings are anagrams.
        
        Args:
             input1 (str): First string
             input2 (str): Second string
        
        Returns:
             bool: True if anagram, False otherwise
        
        Raises:
             TypeError: If inputs are not strings
             ValueError: If inputs are empty
        """
        
        if not isinstance(input1,str) or not isinstance(input1,str):
            raise TypeError("TypeError")
                
        input1 = input1.strip().lower()
        input2 = input2.strip().lower()
        
        if sorted(input1) == sorted(input2):
            return "Anagram"
        else:
            return "Not anagram"

if __name__ == "__main__":
    try:
        input1 = input("Enter a string 1: ")
        input1 = input1.strip()
        input2 = input("Enter a string 2: ")
        input2 = input2.strip()
        obj = Anagram()
        result = obj.Anagram_check(input1,input2)
        print(f"Both strings are: {result} ")
        
    except TypeError as e:
        print(e)
    except ValueError as e:
        print(e)
    except Exception as e:
        print("unexpected error",e)
```
