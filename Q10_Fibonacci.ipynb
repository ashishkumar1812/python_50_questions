```python
def fibonacci(num: int) -> list:
    """
    This function gives the fibonacci series upto given number.
    Args:
        Takes a integer as an input.
    Return:
        returns a list of fibonacci series
    raises:
        TypeError and ValueError
    """
    if not isinstance(num,int):
        raise TypeError("Input must be an integer.")
    if num <= 0:
        raise ValueError("Input must not be empty.")
        
    a, b = 0, 1
    list1 = []
    for i in range(num):
        list1.append(a)
        a, b = b, a + b
    return list1

# main execution
if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number: "))
        result = fibonacci(user_input)
        print(f"Fibonacci series upto number {user_input} is {result}")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print("something went wrong",e)
        
```
