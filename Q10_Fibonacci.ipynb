```python
class Fibonacci:
    def __init__(self):
        print("This class generates Fibonacci series up to a given number.")

    def fibonacci_series(self, num: int) -> list:
        """
        This method returns the Fibonacci series up to a given number.

        Args:
            num (int): Number of terms.

        Returns:
            list: Fibonacci series as a list.

        Raises:
            TypeError: If input is not an integer.
            ValueError: If input is less than or equal to 0.
        """
        # type check
        if not isinstance(num, int):
            raise TypeError("Input must be an integer.")

        # value check
        if num <= 0:
            raise ValueError("Input must be greater than 0.")

        a, b = 0, 1
        series = []

        for _ in range(num):
            series.append(a)
            a, b = b, a + b

        return series


# main execution
if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number: "))

        obj = Fibonacci()
        result = obj.fibonacci_series(user_input)

        print(f"Fibonacci series upto {user_input} terms is: {result}")

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
