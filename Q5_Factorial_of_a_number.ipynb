```python
class Factorial:
    def __init__(self):
        print("This class calculates factorial of a number.")

    def factorial(self, number: int) -> int:
        """
        Calculate the factorial of a number.

        Args:
            number (int): A positive integer.

        Returns:
            int: Factorial of the given number.

        Raises:
            TypeError: If input is boolean or not an integer.
            ValueError: If number is negative or zero.
        """
        # boolean check
        if isinstance(number, bool):
            raise TypeError("Input is a boolean (not allowed).")

        # integer check
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")

        # negative / zero check
        if number < 1:
            raise ValueError("Number must be a positive integer.")

        fact = 1
        for i in range(1, number + 1):
            fact *= i

        return fact


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a number: ").strip()

        if not user_input.isdigit():
            raise TypeError("Input must be a positive integer.")

        number = int(user_input)

        obj = Factorial()
        result = obj.factorial(number)

        print(f"Factorial of {number} is: {result}")

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
