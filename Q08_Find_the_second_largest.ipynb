```python
class SecondLargest:
    def __init__(self):
        print("This class finds the second largest element in a list.")

    def second_largest_element(self, user_input: list) -> float:
        """
        This method finds the second largest element in a given list.

        Args:
            user_input (list): List of numeric values.

        Returns:
            float: Second largest element.

        Raises:
            TypeError: If input is not a list or contains non-numeric elements.
            ValueError: If list is empty or has less than two elements.
        """
        # list check
        if not isinstance(user_input, list):
            raise TypeError("Input must be a list.")

        # empty list check
        if not user_input:
            raise ValueError("List must not be empty.")

        # list length check
        if len(user_input) < 2:
            raise ValueError("List must contain at least two elements.")

        # element type check
        for item in user_input:
            if not isinstance(item, (int, float)):
                raise TypeError("All elements must be numbers.")

        # logic
        largest = second_largest = float('-inf')

        for item in user_input:
            if item > largest:
                second_largest = largest
                largest = item
            elif largest > item > second_largest:
                second_largest = item

        if second_largest == float('-inf'):
            raise ValueError("No second largest element found (all values may be equal).")

        return second_largest


# main execution
if __name__ == "__main__":
    try:
        user_input = list(map(float, input("Enter elements separated by spaces: ").split()))

        obj = SecondLargest()
        result = obj.second_largest_element(user_input)

        print(f"{result} is the second largest element in the given list.")

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
