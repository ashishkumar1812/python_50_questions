```python
class ReverseString:
    def __init__(self):
        print("This class reverses a string using a for loop.")

    def reverse_a_string(self, given_string: str) -> str:
        """
        This method reverses a string using a for loop.
        
        Args:
            given_string (str): The string to be reversed.
        
        Returns:
            str: Reversed string.
        
        Raises:
            TypeError: If input is not a string.
        """
        # validate input type
        if not isinstance(given_string, str):
            raise TypeError("Input must be a string.")

        reverse_string = ""
        for ch in given_string:
            reverse_string = ch + reverse_string

        return reverse_string


# main execution
if __name__ == "__main__":
    try:
        # Taking input from the user
        given_string = input("Enter the string: ")

        # Creating object
        obj = ReverseString()

        # Calling method
        result = obj.reverse_a_string(given_string)

        # Printing output
        print(f"Reversed String: {result}")

    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
