```python
class LeapYear:
    def __init__(self):
        print("This is LeapYear class")
    def leap_year_check(self,year: int) -> bool:
        """
        This method checks if given year is leap year or not.
        Args:
            Year = YYYY format
        Returns:
            Boolean: True or False
        Raises:
            TypeError
        """
        if not isinstance(year, int):
            raise TypeError
        if year <= 0:
            return False
        if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
            return True
        else:
            return False
# main execution
if __name__ == "__main__":
    try:
        year = int(input("Enter a year in YYYY format: "))
        obj = LeapYear()
        result = obj.leap_year_check(year)
        print(f"{year} is a leap year: {result}")
    except TypeError as e:
        print("TypeError: Year must be an integer")
    except ValueError as e:
        print("ValueError",e)
    except Exception as e:
        print("something went wrong",e)

```
