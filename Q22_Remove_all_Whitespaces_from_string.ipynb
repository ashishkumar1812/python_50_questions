```python
class RemoveWhitespace:
    def __init__(self):
        print("This class removes all whitespace from a given string.")

    def remove_whitespace(self, given_string: str) -> str:
        """
        This method removes all whitespace characters from a string.

        Args:
            given_string (str): Input string

        Returns:
            str: String without any whitespace characters

        Raises:
            TypeError: If input is not a string
        """
        # validate input type
        if not isinstance(given_string, str):
            raise TypeError("Input must be a string.")

        result_string = ""

        for ch in given_string:
            # skip whitespace characters
            if not ch.isspace():
                result_string += ch

        return result_string


# main execution
if __name__ == "__main__":
    try:
        # taking input from user
        user_input = input("Enter a string: ")

        # creating object
        obj = RemoveWhitespace()

        # calling method
        result = obj.remove_whitespace(user_input)

        # printing output
        print("String after removing whitespace:")
        print(result)

    except TypeError as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
