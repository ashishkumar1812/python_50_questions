```python
class DigitStringChecker:
    def __init__(self):
        print("This class checks if a string contains only numeric characters.")

    def is_only_digits(self, string: str) -> bool:
        """
        Check if the given string contains only digits.

        Args:
            string (str): Input string

        Returns:
            bool: True if string contains only digits, False otherwise

        Raises:
            TypeError: If input is not a string
        """
        if not isinstance(string, str):
            raise TypeError("Input must be a string.")

        # Empty string is not considered numeric
        if not string:
            return False

        for ch in string:
            if not ch.isdigit():
                return False
        return True


# main execution
if __name__ == "__main__":
    try:
        string_input = input("Enter a string: ").strip()

        obj = DigitStringChecker()
        result = obj.is_only_digits(string_input)

        if result:
            print(f"The string '{string_input}' contains only digits.")
        else:
            print(f"The string '{string_input}' does NOT contain only digits.")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
