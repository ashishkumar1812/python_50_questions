```python
class SumOfSquares:
    def __init__(self):
        print("This class calculates the sum of squares of the first n natural numbers.")

    def calculate(self, n: int) -> int:
        """
        Calculate sum of squares of first n natural numbers.

        Args:
            n (int): Number of natural numbers

        Returns:
            int: Sum of squares

        Raises:
            TypeError: If input is not an integer
            ValueError: If n is less than 1
        """
        if not isinstance(n, int):
            raise TypeError("Input must be an integer.")
        if n < 1:
            raise ValueError("Input must be a positive integer.")

        total = 0
        for i in range(1, n + 1):
            total += i * i  # square and add
        return total


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a positive integer n: ").strip()

        if not user_input.isdigit():
            raise TypeError("Input must be a positive integer.")

        n = int(user_input)

        obj = SumOfSquares()
        result = obj.calculate(n)

        print(f"Sum of squares of first {n} natural numbers: {result}")

    except (TypeError, ValueError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
