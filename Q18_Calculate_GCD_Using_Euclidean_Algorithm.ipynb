```python
class GCD:
    def __init__(self):
        print("This class calculates GCD using Euclidean Algorithm")

    def find_gcd(self, a: int, b: int) -> int:
        """
        Finds GCD of two integers.
        Args:
            a (int): First number
            b (int): Second number
        Returns:
            int: GCD of a and b
        Raises:
            TypeError and ValueError
        """
        if not isinstance(a, int) or not isinstance(b, int):
            raise TypeError("Inputs must be integers.")

        if a == 0 and b == 0:
            raise ValueError("Both numbers cannot be zero.")

        a, b = abs(a), abs(b)

        while b != 0:
            a, b = b, a % b

        return a


if __name__ == "__main__":
    try:
        x = int(input("Enter first number: "))
        y = int(input("Enter second number: "))

        obj = GCD()
        result = obj.find_gcd(x, y)
        print(f"GCD of {x} and {y} is: {result}")
        
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
