```python
class Queue:
    def __init__(self):
        """
        Initialize an empty queue.
        """
        print("This class implements a queue using a list (FIFO).")
        self.items = []

    def enqueue(self, item):
        """
        Add an element to the end of the queue.

        Args:
            item: Any value to be added to the queue.
        """
        self.items.append(item)
        print(f"Enqueued: {item}")

    def dequeue(self):
        """
        Remove and return the front element of the queue.

        Returns:
            The front element if queue is not empty, otherwise None.
        """
        if self.items:
            dequeued_item = self.items.pop(0)
            print(f"Dequeued: {dequeued_item}")
            return dequeued_item
        else:
            print("Queue is empty. Cannot dequeue.")
            return None

    def peek(self):
        """
        Return the front element without removing it.

        Returns:
            The front element if queue is not empty, otherwise None.
        """
        if self.items:
            front_item = self.items[0]
            print(f"Front element: {front_item}")
            return front_item
        else:
            print("Queue is empty. Nothing to peek.")
            return None

    def is_empty(self):
        """
        Check if the queue is empty.

        Returns:
            bool: True if queue is empty, False otherwise.
        """
        return len(self.items) == 0


# main execution
if __name__ == "__main__":
    queue = Queue()

    # enqueue elements
    queue.enqueue(10)
    queue.enqueue(20)
    queue.enqueue(30)

    # peek at front
    queue.peek()  # Expected Output: 10

    # dequeue elements
    queue.dequeue()  # Expected Output: 10
    queue.dequeue()  # Expected Output: 20
    queue.peek()     # Expected Output: 30
    queue.dequeue()  # Expected Output: 30
    queue.dequeue()  # Expected Output: None (queue empty)
```
