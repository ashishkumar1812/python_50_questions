```python
class SubsequenceChecker:
    def __init__(self):
        print("This class checks if a smaller list is a subsequence of a larger list.")

    def is_subsequence(self, larger_list: list, smaller_list: list) -> bool:
        """
        Check if smaller_list is a subsequence of larger_list.

        Args:
            larger_list (list): The list in which to search
            smaller_list (list): The list to check as subsequence

        Returns:
            bool: True if smaller_list is a subsequence of larger_list, False otherwise

        Raises:
            TypeError: If inputs are not lists
        """
        if not isinstance(larger_list, list) or not isinstance(smaller_list, list):
            raise TypeError("Both inputs must be lists.")

        if not smaller_list:  # empty list is always a subsequence
            return True

        index_small = 0
        for item in larger_list:
            if item == smaller_list[index_small]:
                index_small += 1
                if index_small == len(smaller_list):
                    return True

        return False


# main execution
if __name__ == "__main__":
    try:
        raw_larger = input("Enter the larger list (e.g., [1,3,5,7,9]): ").strip()
        larger_list = eval(raw_larger)

        raw_smaller = input("Enter the smaller list (e.g., [3,7]): ").strip()
        smaller_list = eval(raw_smaller)

        obj = SubsequenceChecker()
        result = obj.is_subsequence(larger_list, smaller_list)

        if result:
            print(f"{smaller_list} is a subsequence of {larger_list}.")
        else:
            print(f"{smaller_list} is NOT a subsequence of {larger_list}.")

    except (TypeError, SyntaxError) as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
