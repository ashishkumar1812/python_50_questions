```python
# Count Occurrences of Each Character in a String
char_count = {}
text = "karan"
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

```python
class CharacterCount:
    def __init__(self):
        print("This class counts occurrences of each character in a string.")

    def count_characters(self, text: str) -> dict:
        """
        This method counts occurrences of each character in a string.
        
        Args:
            text (str): Input string
        
        Returns:
            dict: Dictionary with character counts
        
        Raises:
            TypeError: If input is not a string
            ValueError: If input is empty
        """
        if not isinstance(text, str):
            raise TypeError("Input must be a string.")
        
        text = text.strip()
        if not text:
            raise ValueError("Input string must not be empty.")

        char_count = {}

        for ch in text:
            if ch in char_count:
                char_count[ch] += 1
            else:
                char_count[ch] = 1

        return char_count


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter a string: ")
        obj = CharacterCount()
        result = obj.count_characters(user_input)
        print("Character occurrences:", result)
    except TypeError as e:
        print("TypeError:", e)
    except ValueError as e:
        print("ValueError:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
