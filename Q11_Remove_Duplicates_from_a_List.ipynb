```python
def remove_duplicates(list1: list) -> list:
    """
    This function removes duplicate elements from a list.

    Args:
        list1 (list): List containing duplicate elements.

    Returns:
        list: List without duplicate values.

    Raises:
        TypeError: If input is not a list.
        ValueError: If list is empty.
    """
    # type check
    if not isinstance(list1, list):
        raise TypeError("Input must be a list")

    # empty check
    if not list1:
        raise ValueError("List must not be empty")

    unique_list = []

    for item in list1:
        if item not in unique_list:
            unique_list.append(item)

    return unique_list


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter elements separated by spaces: ").split()
        result = remove_duplicates(user_input)
        print(f"List without duplicates: {result}")

    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Something went wrong:", e)
```
