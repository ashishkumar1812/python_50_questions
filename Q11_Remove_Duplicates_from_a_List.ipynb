```python
def remove_duplicates(list1: list) -> list:
    """
    This function removes duplicates from a list.
    Args:
        Takes a list as an arguments.
    Returns:
        returns list without any duplicate values.
    Raises:
        TypeError: Input must be a list.
        ValueError: list should not be empty.
    """
    if not isinstance(list1,list):
        raise TypeError("Input should be a list")
    if not list1:
        raise ValueError("List should not be empty.")
    list2 = []
    for i in list1:
        if i not in list2:
            list2.append(i)
    return list2

# main execution
if __name__ == "__main__":
    try:
        user_input = list(map(str, input("Enter the list seperated by spaces: ").split(" ")))
        result = remove_duplicates(user_input)
        print(f"list without duplicates: {result}")
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("something went wrong",e)
```
