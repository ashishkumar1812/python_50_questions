```python
class ReverseInteger:
    def __init__(self):
        print("This class reverses the digits of an integer.")

    def reverse(self, number: int) -> int:
        """
        Reverse the digits of the given integer.

        Args:
            number (int): Integer to reverse

        Returns:
            int: Reversed integer

        Raises:
            TypeError: If input is not an integer
        """
        if not isinstance(number, int):
            raise TypeError("Input must be an integer.")

        is_negative = number < 0
        num_str = str(abs(number))
        reversed_str = num_str[::-1]  # reverse the string
        reversed_number = int(reversed_str)

        if is_negative:
            reversed_number = -reversed_number

        return reversed_number


# main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter an integer to reverse: ").strip()

        if user_input.startswith('-'):
            if not user_input[1:].isdigit():
                raise TypeError("Input must be a valid integer.")
        else:
            if not user_input.isdigit():
                raise TypeError("Input must be a valid integer.")

        number = int(user_input)

        obj = ReverseInteger()
        result = obj.reverse(number)

        print(f"Reversed integer: {result}")

    except TypeError as e:
        print("Error:", e)
    except Exception as e:
        print("Something went wrong:", e)
```
